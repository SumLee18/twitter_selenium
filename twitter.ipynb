# Capture replies for each post

```python
def copy_list(list):
    list2 = []
    for item in list:
        list2.append(item)
    return list2
```

```python
post_dict = {}
reply_time = []
reply_account_tag = []
reply_account_name_and_tag = []
reply_account_name = [] 
reply_content = []
reply_num = []
Text = ""
num_text = ""

for num in link_dict.keys():
    all_reply = all_reply_dict[num]
    for i in range(len(all_reply)):
        efg = all_reply[i].find_all('div', class_="css-175oi2r r-18u37iz r-1q142lx")
        for reply in efg:
            if reply.find('a', href=True):
                abc = reply.find('a', href=True)['href']
                hijk = re.findall(r'/([^/]+)/', abc)
                reply_account_tag.append("@"+hijk[0])
                date = reply.find('a', href=True).find('time').text
                '''
                if len(date) <= 3:
                    today = datetime.now()
            
                elif len(date) <= 6:
                    date = date + ", 2025"
                '''
                #print(date)
                reply_time.append("@"+hijk[0]+" "+date)

        index_list = []
        #print(reply.find("div"))
        lmn = all_reply[i].find_all('span', class_="css-1jxf684 r-bcqeeo r-1ttztb7 r-qvutc0 r-poiln3")
        for j in range(len(lmn)):
            #print(lmn[j])
            if lmn[j].text in reply_account_tag:
                reply_account_name_and_tag.append(lmn[j-1].text + lmn[j].text) #name@name
                reply_account_name.append(lmn[j-1].text) #name
                index_list.append(j)
                #print(reply)

        text_list = [] 
        for i in index_list:
            name = lmn[i].text
            if name in reply_account_tag:
                j = i+1 
                if lmn[j].text == "·":
                    j = j + 1
                    w = 0
                    while True:
                        w = w + 1
                        if lmn[j].text.isdigit() or lmn[j].text == "Relevant people":
                            while lmn[j].text.isdigit():
                                num_text = num_text + lmn[j].text #
                                j=j+1
                            reply_num.append(name+" "+num_text)
                            for l in text_list:
                                Text = Text + l
                            reply_content.append(Text + " + " + name)
                            Text = ""
                            num_text =""
                            text_list.clear()
                            break
                        elif lmn[j].text in reply_account_tag:
                            reply_num.append(name+" "+num_text)
                            for l in range(w-2):
                                Text = Text + text_list[l]
                            reply_content.append(Text + " + " + name)
                            Text = ""
                            num_text=""
                            text_list.clear()
                            break

                        text_list.append(lmn[j].text)
                        j = j + 1 
    replytime = copy_list(reply_time)
    replyaccount_name_and_tag = copy_list(reply_account_name_and_tag)
    replyaccount_name = copy_list(reply_account_name)
    replyaccount_tag = copy_list(reply_account_tag)
    replynum = copy_list(reply_num)
    replycontent = copy_list(reply_content)
    post_dict[num] = [replytime, replyaccount_name_and_tag, replyaccount_name, replyaccount_tag, replynum, replycontent,link_dict[num]]

    print(reply_account_name_and_tag)
    print(len(reply_num))
    print(len(reply_time))

    reply_time.clear()
    reply_account_name_and_tag.clear()
    reply_account_name.clear()
    reply_account_tag.clear()
    reply_num.clear()
    reply_content.clear()
```

```python
def find_first_occurrence(text, string_list):
    for string in string_list:
        if string in text:
            return string  # Return the first found string
    return None  # Return None if no strings are found

def find_first_occurrence_num(text, string_list, name):
    for string in string_list:
        date_index = string.find(" ")
        ad_name = string[0:date_index]
        ad_num = string[date_index+1:]
        if ad_name in name:
            if ad_num in text:
                return ad_num,ad_name  # Return the first found string and its name
    return None

def search_from_bottom(text, string_list, name):
    # Reverse the text
    reversed_name = name[::-1]
    reversed_text = text[::-1]

    for string in string_list:
        string_reversed = string[::-1]
        date_index = string_reversed.find(" ")
        ad_name = string_reversed[date_index+1:]
        ad_num = string_reversed[0:date_index]
        if ad_name in reversed_name:
            if ad_num in reversed_text:
                return len(reversed_text)-2-reversed_text.find(ad_num), ad_num[::-1], ad_name[::-1] 
    return None

def copy_list(list):
    list2 = []
    for item in list:
        list2.append(item)
    return list2
```

```python
post_reply_dict = {}
post_reply_name = []
post_reply = []
post_reply_time = []
post_reply_info = []
post_reply_info_filtered = []

for num in link_dict.keys():
    all_reply = all_reply_dict[num]
    for j in range(len(all_reply)):
        hij = all_reply[j].find_all('div', class_="css-175oi2r")
        for i in range(len(hij)):
            reply_account_name_and_tag = post_dict[num][1]
            for name_index in range(len(reply_account_name_and_tag)):

                index = name_index + 1
                first_index = hij[i].text.find(reply_account_name_and_tag[name_index])
                last_index = -1
                if first_index > -1:
                    found = find_first_occurrence(hij[i].text[first_index+len(reply_account_name_and_tag[name_index]):],reply_account_name_and_tag)
                    if found != None and first_index > -1:
                        last_index = hij[i].text[first_index+len(reply_account_name_and_tag[name_index]):].find(found)
                    elif found == None and first_index > -1:
                        found = "Relevant people"
                        last_index = hij[i].text[first_index+len(reply_account_name_and_tag[name_index]):].find(found) 
                    if last_index == -1:
                        found = ""
                        last_index = len(hij[i].text)

            
    
                if (first_index > -1 and last_index > -1):
                    last_index = last_index + len(hij[i].text[:first_index+len(reply_account_name_and_tag[name_index])])
                    Text = hij[i].text[first_index:last_index]
                    post_reply_info.append(Text)
                    reply_time = post_dict[num][0]
                    ad = find_first_occurrence_num(Text, reply_time,reply_account_name_and_tag[name_index])
                    #if ad == None:
                        #print(hij[i].text)
                        #print("None")
                    if ad != None:
                        ad_time = ad[0]
                        ad_name = ad[1]
                        ad_time_index = Text.find(ad_time)
                        #print(Text)
                        #print(reply_account_name_and_tag[name_index])
                        #print(ad_name)
                        if reply_account_name_and_tag[name_index].find(ad_name) > -1:
                            #if ad_time_index == -1:
                                #print("Hello")
                            if ad_time_index != -1:
                                # print(Text_no_time)
                                Text_no_time = Text[ad_time_index+len(ad_time):]
                                reply_num = post_dict[num][4]
                                ad = search_from_bottom(Text_no_time, reply_num,reply_account_name_and_tag[name_index])
                                if ad != None:
                                    ad_num = ad[1]
                                    ad_name = ad[2]
                                    ad_num_index = ad[0]
                                    #print(reply_account_name_and_tag[name_index])
                                    if ad_num == "":
                                        Text_no_num = Text_no_time
                                    else:
                                        Text_no_num = Text_no_time[:ad_num_index]
                                    post_reply_name.append(reply_account_name_and_tag[name_index])
                                    post_reply_time.append(ad_time)
                                    post_reply.append(Text_no_num)
                                    post_reply_info_filtered.append(Text)
    replyinfo = copy_list(post_reply_info)
    replyinfo_filtered = copy_list(post_reply_info_filtered)
    replyname = copy_list(post_reply_name)
    replytime = copy_list(post_reply_time)
    replytext = copy_list(post_reply)
    replyinfo_filtered = copy_list(post_reply_info_filtered)
    post_reply_dict[num] = [replyinfo,replyinfo_filtered,replyname,replytime,replytext,link_dict[num]]
    post_reply_info.clear()
    post_reply_name.clear()
    post_reply_time.clear()
    post_reply.clear()
    post_reply_info_filtered.clear()      
```
